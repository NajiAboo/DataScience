```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import os


collisionData = pd.read_csv(os.path.join(os.path.dirname("~/Downloads/"),"NYPD_Motor_Vehicle_Collisions.csv"))
collisionData
```

```python
numberInjuries = collisionData[['DATE','TIME','BOROUGH','LOCATION','NUMBER OF PERSONS INJURED','NUMBER OF PERSONS KILLED']]
numberInjuries
```

```python
groupByBorough = numberInjuries.groupby(['BOROUGH'],as_index=False)['NUMBER OF PERSONS INJURED','NUMBER OF PERSONS KILLED'].sum()
print groupByBorough
```

```python
ax = groupByBorough.plot(kind='bar',figsize=(15,8));
ax.set_xticklabels(groupByBorough['BOROUGH'],rotation='horizontal');
ax.set_ylabel("Count");
ax.set_xlabel("Borough");
ax.set_title("Number of Persons Injuried/Killed in vehicle collisions grouped by boroughs");

```

```python
groupByBoroughDate = numberInjuries.groupby(['BOROUGH','DATE'],as_index=False)['NUMBER OF PERSONS INJURED','NUMBER OF PERSONS KILLED'].sum()
print groupByBoroughDate.head()
```
